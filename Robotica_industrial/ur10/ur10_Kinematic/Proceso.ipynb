## Proceso moveit2 robot UR10

#### En primer momento se selecciona el URDF en el asistente de moveit,este URDF se toma del paquete creado del robot UR10 

#### En el asistente de moveit se agregan las colisiones

#### Se agrega el el virtual joint, este se encargara de tomar un link del urdf y dejarlo como punto fijo en el mundo

#### Se agregan las juntas que conformaran el brazo, esto se agregara como un grupo

#### Por consiguiente, se guarda la configuracion del asistente con el nombre de "urconf" 

#### Se ejecuta el colcon build para crear correctamente el paquete generado por el asistente

#### Por ultimo se visualiza el robot en Rviz
